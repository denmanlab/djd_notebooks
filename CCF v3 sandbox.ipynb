```python

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from allensdk.core.mouse_connectivity_cache import MouseConnectivityCache
# pandas for nice tables
import pandas as pd
```

## load the most current CCF from the allensdk

```python
mcc = MouseConnectivityCache(manifest_file='mcc/manifest.json', resolution=10)
```

```python
av, _ = mcc.get_annotation_volume()
```

```python
tv, _ = mcc.get_template_volume()
```

#### the shape of the returned data is a 3D matrix of the brain, nslices x 800 x 1140. in this case, 1320 slices at 10um resolution

```python
av.shape
```

```python
tv.shape
```

### show the annotated structures
#### the color in the below is based on the annoted id; the annotation can be found in structures.csv in wherever the manifest downloaded the data to (in my case, this was the root directory of the notebook server.)

```python
plt.imshow(av[700,:,:],vmin=0, vmax=2000,cmap=plt.cm.prism)
```
