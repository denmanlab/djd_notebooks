```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import djd.generalephys as ephys
import cPickle as pkl
import os, sys,glob, copy
import pandas as pd
import djd.eyetrackingTools as eye

%matplotlib inline

%load_ext autoreload
%autoreload 2
```

```python

path = r'\\Drobo5N\Public\Dan\ephys\multi_area\M192079\eyetracking'
uv_eye = eye.loadEyeTraces(os.path.join(path,'2015-09-02-152853255000-1-output.h5'))
vis_eye = eye.loadEyeTraces(os.path.join(path,'2015-09-02-151819233000-1-output.h5'))
```

```python
data=vis_eye
start = 0
range = len(data['pupil_positions'][:,0])

gs = gridspec.GridSpec(4, 2)
ax1=plt.subplot(gs[0,0])
ax2=plt.subplot(gs[0,1])
ax3=plt.subplot(gs[1:,:])

data=vis_eye
ax1.plot(np.arange(0,len(data['pupil_positions'][:,0][start:start+range])),data['pupil_positions'][:,0][start:start+range],color='#6B8E23',ls='--')
ax1.plot(np.arange(0,len(data['pupil_positions'][:,1][start:start+range])),data['pupil_positions'][:,1][start:start+range],color='#6B8E23')
ax2.plot(np.arange(0,len(data['pupil_area'][start:start+range])),data['pupil_area'][start:start+range],color='#6B8E23')
ax3.scatter(data['pupil_positions'][:,0][start:start+range],data['pupil_positions'][:,1][start:start+range],s=20,alpha=0.5,c=np.abs(np.array(data['pupil_area'][start:start+range])*-1),cmap=plt.cm.Greens,vmin=2000,vmax=20000)#,color='k',ls='-o')


data=uv_eye
ax1.plot(np.arange(0,len(data['pupil_positions'][:,0][start:start+range])),data['pupil_positions'][:,0][start:start+range],color='#4B0082',ls='--')
ax1.plot(np.arange(0,len(data['pupil_positions'][:,1][start:start+range])),data['pupil_positions'][:,1][start:start+range],color='#4B0082')
ax2.plot(np.arange(0,len(data['pupil_area'][start:start+range])),data['pupil_area'][start:start+range],color='#4B0082')
ax3.scatter(data['pupil_positions'][:,0][start:start+range]+0.5,data['pupil_positions'][:,1][start:start+range]+0.5,s=20,alpha=0.5,c=np.abs(np.array(data['pupil_area'][start:start+range])*-1),cmap=plt.cm.Purples,vmin=2000,vmax=20000)#,color='k',ls='-o')

ax1.set_xlabel('time [frame#]')
ax1.set_ylabel('pupil position')

ax2.set_xlabel('time [frame#]')
ax2.set_ylabel('pupil area')

ax3.set_xlabel('pupil position [x]')
ax3.set_ylabel('pupil position [y]')
ax3.set_xlim(60,110)
ax3.set_ylim(120,170)

#plt.tight_layout()
f=plt.gcf();f.set_size_inches(14.5, 14.5)
```
