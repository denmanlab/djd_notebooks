```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
n = 10000
```

```python
noise_cors = np.array(np.random.normal(0,1,10000)/8.5+0.1)
signal_cors =noise_cors + np.random.normal(0,1,10000)/4.
```

```python
plt.plot(signal_cors,noise_cors,'o')
plt.xlim(-1,1)
plt.ylim(-1,1)
```

```python
sns.set_style('white')
```

```python
a=sns.kdeplot(signal_cors,noise_cors,cmap=plt.cm.plasma)
plt.gca().set_aspect(1)
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel('signal correlation',size=14)
plt.ylabel('noise correlation',size=14)
plt.gca().set_yticks([-1,-.5,0,.5,1])
plt.gcf().savefig('/Users/danieljdenman/Dropbox/allenTransfer/cors.eps',format='eps')
plt.gcf().savefig('/Users/danieljdenman/Dropbox/allenTransfer/cors.png',format='png',dpi=300)
```
