```python
def bin_data(x,y,bins):
    mean=np.zeros(len(bins))
    std =np.zeros(len(bins))
    se =np.zeros(len(bins))
    binsize = np.diff(bins)[0]
    for i,bin_ in enumerate(bins):
        data = y[np.where((x > (bin_ - binsize/2)) & (x < (bin_ + binsize/2)))[0]]
        mean[i] = np.mean(data)
        std[i] =  np.std(data)
        se[i] =  np.std(data)/np.sqrt(len(data))
    return mean,std,se
```
