```python
import djd.generalephys as ephys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
uvpath = r'/Volumes/Public/Dan/dome/lag_test_2015-09-03_09-03-20_uv'
greenpath = r'/Volumes/Public/Dan/dome/lag_test_2015-09-03_09-05-11_vis'
```

```python
uv = ephys.oe.loadFolder(uvpath)
```

```python
green = ephys.oe.loadFolder(greenpath)
```

```python
green['times']=plt.mlab.cross_from_below(green['100_PAI7']['data'],1)
uv['times']=plt.mlab.cross_from_below(uv['100_PAI7']['data'],1)
```

```python
green['bright']=green['times'][1::2]
green['dark']=green['times'][2::2]

uv['bright']=uv['times'][1::2]
uv['dark']=uv['times'][2::2]
```

```python
number = 1

plt.plot(green['100_PAI5']['data'],color='#6B8E23')
plt.plot(green['times'],np.ones(len(green['times'])),'go')
plt.xlim(green['bright'][number]-250,green['bright'][number]+2250)

plt.figure()
plt.plot(uv['100_PAI5']['data'],color='#4B0082')
plt.plot(uv['times'],np.ones(len(uv['times']))-0.1,'bo')
plt.xlim(uv['bright'][number]-250,uv['bright'][number]+2250)
```
